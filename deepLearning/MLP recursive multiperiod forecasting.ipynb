```python
from keras.models import Sequential
from keras.layers import Dense
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from master_function import data_preprocessing, plot_train_test_values
from master_function import recursive_mpf
from master_function import calculate_directional_accuracy
from sklearn.metrics import mean_squared_error
```

```python
# Importing the data
data = np.reshape(np.array(pd.read_excel('Temperature_Basel.xlsx').dropna()), (-1))
data
```

```python
#гиперпараметры
num_lags = 500
train_test_split = 0.8
num_neurons_in_hidden_layers = 100
num_epochs = 200
batch_size = 12
```

```python
#train test split
x_train, y_train, x_test, y_test = data_preprocessing(data, num_lags, train_test_split)
```

```python
#mlp архитектруа keras
model = Sequential()
model.add(Dense(num_neurons_in_hidden_layers, input_dim = num_lags, activation = 'relu'))
model.add(Dense(num_neurons_in_hidden_layers, activation = 'relu'))
model.add(Dense(num_neurons_in_hidden_layers, activation = 'relu')) 
model.add(Dense(num_neurons_in_hidden_layers, activation = 'relu'))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
#4 hidden layers+ output layer and compiling
```

```python
#теперь fit
model.fit(x_train, np.reshape(y_train, (-1, 1)), epochs = num_epochs, batch_size = batch_size)
```

```python
#предсказания in sample
y_predicted_train = np.reshape(model.predict(x_train), (-1, 1))
#предсказания oos test set - recursive basis
x_tets, y_predicted = recursive_mpf(x_test, y_test, num_lags, model, architecture = 'MLP')
# recursive_mpf() function takes the following arguments:
#1_test set features that will continuously be updated. They are represented by the variable x_test.
#2_test set dependent variables. They are represented by the variable y_test.
#3_number of lags. This variable is represented by num_lags.
#4_fitted model as defined by the variable model.
#5_architecture:  MLP for two-dimensional arrays or LSTM for three-dimensional arrays.
```
